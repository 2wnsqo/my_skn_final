```python
import json
import matplotlib.pyplot as plt

# 1. JSON 파일 로드
with open("ai_hub_scored_results.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 2. score 값만 추출
scores = [item["score"] for item in data]

# 3. 히스토그램 시각화 (5점 단위 bin)
plt.figure(figsize=(10, 6))
plt.hist(scores, bins=range(0, 105, 5), edgecolor='black')  # 0~100까지 5점 간격
plt.title("Score Distribution (5-point intervals)")
plt.xlabel("Score")
plt.ylabel("Number of Samples")
plt.xticks(range(0, 105, 5))
plt.grid(True)
plt.show()
```
